```python
import pandas as pd
```

```python
df = pd.read_excel("sfdc_final.xlsx")
```

```python
df
```

```python
sfdc_group = df.groupby(['NAICS']).describe()
# sfdc_group = pd.DataFrame(sfdc_group.first())
```

```python
avg_price = pd.DataFrame(sfdc_group["Total Price (converted)"]["mean"])
```

```python
avg_price
```

```python
company_df = pd.read_excel("NAICS_total_employee.xlsx")
```

```python
company_df
```

```python
merge_df = pd.merge(company_df, avg_price, left_on="NAICS Code", right_on="NAICS", how='left')
```

```python
merge_df
```

```python
merge_df["mean"] = merge_df["mean"].fillna(merge_df['mean'].mean())
```

```python
merge_df
```

```python
merge_df.to_excel("merge_df.xlsx")
```
