```python
# Agricultural Chemicals
Agricultural_Chemicals = sfdc_pricing.loc[sfdc_pricing["Sub-Vertical (Reporting)"] == "Agricultural Chemicals"]
Agricultural_Chemicals= Agricultural_Chemicals.dropna()
Agricultural_Chemicals
```

```python
sub_vertical_list = list(pd.unique(sfdc_pricing["Sub-Vertical (Reporting)"]))
sub_vertical_list
```

```python
d = {}
for i in sub_vertical_list:
    df = sfdc_pricing.loc[sfdc_pricing["Sub-Vertical (Reporting)"] == i]
    df = df.dropna()
    key = str(i)
    d[key] = list(pd.unique(df["Protection Need"]))
```

```python
print(d)
```

```python
protection_need_for_sub_ver = pd.DataFrame(d.items(), columns = ['Sub Verticals', 'Protection Needs'])
protection_need_for_sub_ver 
```

```python
df = pd.DataFrame(columns = ["Number of Protection Needs"])
df
```

```python
for i in range(len(protection_need_for_sub_ver)):
    length = len(protection_need_for_sub_ver.iloc[i,1])
    df = df.append({"Number of Protection Needs": length},ignore_index = True)
```

```python
df
```

```python
protection_need_for_sub_ver = pd.concat([protection_need_for_sub_ver, df],axis = 1)
protection_need_for_sub_ver
```

```python
protection_need_for_sub_ver.to_excel("protection_need_for_sub_ver.xlsx", index = None)
```
